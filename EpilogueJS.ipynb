```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from config1 import user
from config1 import password1

from sqlalchemy import create_engine
```

```python
# Create engine, connect to DB
engine = create_engine(f'postgresql://{user}:{password1}@localhost:5432/SQLHomework')
connection = engine.connect()
```

```python
# Read in salaries. 
salaries = pd.read_sql("SELECT * FROM salaries", connection)
salaries.head()
```

```python
#Read in titles. 
titles = pd.read_sql("SELECT * FROM titles", connection)

titles.head()
```

```python
# Merge titles and salary using Pandas. 

title_salary = pd.merge(titles, salaries, on='emp_no', how = 'inner')
title_salary.head(15)

#Merged table is odd. Employee 1001 has two different end dates (one that indicates that he or she is still with the company
# and another that shows an end date (for a salary level). Other employees that moved into different roles have new start and 
# end dates. However, their salaries did not always change when they changed positions: see employee 10004. Employee 1007
# seems to have received a promotion from Staff to Senior Staff, but he or she received no increase in salary. The start dates
# and end dates in many of the records seem to differ in the titles file and the salary file. I do hope that employees aren't 
# getting additional responsibilities without getting additional pay. 
```

```python
# Try to query using SQL. 
# My first attempt at this was to do a merge table, but the plot didn't work. I thought it might
# if I read in a query. It didn't, but I thought it was pretty cool to use the join query as a db, so 
# I left it in for posterity. 
joined_TS_db = pd.read_sql('SELECT ttl.emp_no, ttl.title, sal.salary FROM titles AS ttl INNER JOIN salaries as sal ON (ttl.emp_no = sal.emp_no)', connection)
```

```python
# Print to make sure it worked. 
joined_TS_db.head(15)
```

```python
#Group by, and while you're at it, create a calculated column. 

grp_jnd_TS = pd.DataFrame(joined_TS_db.groupby("title").mean()["salary"]).reset_index()
```

```python
grp_jnd_TS
```

```python
plot_df = grp_jnd_TS[["title", "salary"]]
```

```python
# Make a DF just for the plot. (Possibly not necessary; was trying to get around an earlier error.)
plot_df
```

```python
## Make a bar graph. Note: don't try to just do a plt; there will be much whining about a tuple object.
ax = plot_df.salary.plot(kind='bar')
ax.set_xticklabels(plot_df.title, rotation = 45, ha = "right")
ax.set_title("Average Salary by Title")
ax.set_ylabel("Average Salary in $")
plt.tight_layout()
ax.plot()
plt.savefig("AverageSalarybyTitle.png")
```
